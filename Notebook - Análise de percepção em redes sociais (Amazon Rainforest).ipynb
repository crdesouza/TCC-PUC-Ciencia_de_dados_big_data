### 2. Processamento dos dados

### 2.1 Limpeza e padronizao de dados

Nesta etapa ser feita a remoo de textos indesejados, que podem influenciar a anlise das informaes.

i)	Remoo de hashtags;

ii)	Remoo de palavras com @ (menes); 

iii)	Remoo de https:// e http://;

iv)	 Remoo de caracteres grficos como emotions;

v)	Remoo de pontuao.

vi)	Remoo de caracteres de controle;

vii)	Remoo de nmeros;

viii)	Substituio de _ por espao;

ix)	Remoo espaos excessivos e desnecessrios;

x)	Remoo de quebras de linha.


```r

data $text <- gsub("#([a-z|A-Z|0-9|_])*","", data$text) # remover hashtags
data $text <- gsub('@([a-z|A-Z|0-9|_])*', '', data$text) # remover palavras com @ (menes)
data $text <- gsub('https://','', data$text) # remover https://
data$text <- gsub('http://','', data$text) # remover http://
data$text <- gsub('[^[:graph:]]', ' ', data$text) # remove caracteres grficos como emotions
data$text <- gsub('[[:punct:]]', '', data$text) # remove pontuao
data$text <- gsub('[[:cntrl:]]', '', data$text) # remove caracteres de controle
data$text <- gsub("\\w*[0-9]+\\w*\\s*", "", data$text) # remove nmeros
data$text <- gsub('_', ' ', data$text) # substitui sublinhado por espao
data$text = gsub('[ \t]{2,}', ' ', data$text) ## remove espaos desnecessrios
data$text = gsub('^\\s+|\\s+$', ' ', data$text)   ## remove espaos desnecessrios
data$text <- tolower(data$text) # transforma tudo em minor case
data$text  = gsub('\\n', ' ', data$text ) ### remove quebras de linha

head(data) ## mostra um geral dos dados
```

Como pode ser observado, agora os dados textuais tem somente os textos apresentados, sem pontos ou caracteres especiais.

### 2.4 Tokenization

Para prosseguir com as anlises de interesse, foi realizado o processo chamado de "tokenization", que no portugus pode ser traduzido como "tokenizao" ou "segmentao". Esse processo consiste no particionamento dos textos das publicaes em vrios outros registros contendo cada uma das suas palavras componentes. Ou seja, se uma publicao tem 5 palavras, cada palavra passar a contar como uma nova linha na base de dados, porm contendo o registro do seu tweet de origem. Esta  uma etapa de muita importncia para avaliar as informaes das palavras componentes nas etapas seguintes.

A figura abaixo traz um exemplo de tokenization aplicado a uma frase.

Uma etapa importante no processamento de dados textuais consiste na remoo de palavras que no contribuem com anlise de contedo, tais como preposies, conjunes e artigos. Estas palavras so chamadas de "stop-words" e so registradas em dicionrios, que so tomados como base para a remoo das palavras nos contedos textuais. Assim, os dados de interesse so cruzados com estes dicionrios, de forma a remover aquelas palavras identificadas como "stop-words".

### 3.5 Anlise de regras de associao

Com base nos dados de ocorrncia das palavras nas publicaes, tambm foi realizada a anlise de regras de associao (ARA), com o objetivo de avaliar se existem relaes consistentes entre palavras nas publicaes. A anlise foi feita utilizando como critrios de seleo:

- suporte igual ou superior a 0.01875117 (1.875117 %), que corresponde  ocorrncia da espcie no LER em pelo menos 1000 publicaes; 

- confiana de 0,8 para a regra, em que as palavras em LDR devem estar relacionadas com as palavras do LER em pelo menos 80% de suas ocorrncias;

- valor de p de Fisher significativo (<0,05). 

O fluxo de anlise est apresentado abaixo.



```r
## carregamento de pacotes especficos
library(arules)  
library(arulesViz)

## exportao de um objeto no formato desejado para anlise
write.csv(select(tidy_text,tweet_id,word),"word_tweet.csv")


########### leitura deste arquivo de transaes
trans = read.transactions("word_tweet.csv", h=T, sep = ",",format = "single",cols=c("tweet_id","word"))

# Obteno das regas de interesse
rules<- apriori(trans, parameter = list(supp = 0.01875117, conf = 0.8, target="rules", minlen=2,maxlen=2))

########## Obter medidas de interesse para as regras
measures<-interestMeasure(rules,
                               measure=c("coverage","fishersExactTest"),
                               transactions = trans)

############ Adicionar a medida de interesse ao objeto de regras
rules@quality$fisher_p<-measures$fishersExactTest

rules
```

```r
## Resultados gerais das regras
rules %>%
  sort(., by = "lift") %>%           
  head(., n =75 ) %>%                       
  inspect(.)
```

```r
######### fazer um plot para as regras obtidas


plot(rules, method="graph", main="Principais associes",
     alpha=1,cex=1,labelCol="#000000",precision=3, engine="igraph",arrowSize=1)
```

       Anlise resultou em 75 regras entre pares de termos, compostas por 30 palavras que se relacionam entre si. Parte destas palavras esto presentes nas 20 mais importantes apresentadas anteriormente (como carbono e fire), enquanto outras no foram citadas anteriormente (como ice e heatwaves). Dentro do contexto avaliado de percepo ambiental, chama a ateno dentre as regras obtidas as associaes consistentes entre unnatural e heatwaves, dying e fire; emitts e absorbs; peoples e indigenous; carbon e dioxide e absorb. Todas estas associaes indicam que a ocorrncia de uma palavra na publicao est fortemente relacionada a ocorrncia da outra, em uma relao forte de coexistncia. Tambm podem ser observadas outras associaes consistentes de acordo com os critrios considerados, mas que no so relevantes dentro do contexto avaliado de percepo ambiental sobre a regio amaznica.
