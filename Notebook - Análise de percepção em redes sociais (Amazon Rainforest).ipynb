### 2. Processamento dos dados

### 2.1 Limpeza e padronização de dados

Nesta etapa será feita a remoção de textos indesejados, que podem influenciar a análise das informações.

i)	Remoção de hashtags;

ii)	Remoção de palavras com @ (menções); 

iii)	Remoção de “https://” e “http://”;

iv)	 Remoção de caracteres gráficos como emotions;

v)	Remoção de pontuação.

vi)	Remoção de caracteres de controle;

vii)	Remoção de números;

viii)	Substituição de “_” por espaço;

ix)	Remoção espaços excessivos e desnecessários;

x)	Remoção de quebras de linha.


```r

data $text <- gsub("#([a-z|A-Z|0-9|_])*","", data$text) # remover hashtags
data $text <- gsub('@([a-z|A-Z|0-9|_])*', '', data$text) # remover palavras com @ (menções)
data $text <- gsub('https://','', data$text) # remover https://
data$text <- gsub('http://','', data$text) # remover http://
data$text <- gsub('[^[:graph:]]', ' ', data$text) # remove caracteres gráficos como emotions
data$text <- gsub('[[:punct:]]', '', data$text) # remove pontuação
data$text <- gsub('[[:cntrl:]]', '', data$text) # remove caracteres de controle
data$text <- gsub("\\w*[0-9]+\\w*\\s*", "", data$text) # remove números
data$text <- gsub('_', ' ', data$text) # substitui sublinhado por espaço
data$text = gsub('[ \t]{2,}', ' ', data$text) ## remove espaços desnecessários
data$text = gsub('^\\s+|\\s+$', ' ', data$text)   ## remove espaços desnecessários
data$text <- tolower(data$text) # transforma tudo em minor case
data$text  = gsub('\\n', ' ', data$text ) ### remove quebras de linha

head(data) ## mostra um geral dos dados
```

Como pode ser observado, agora os dados textuais tem somente os textos apresentados, sem pontos ou caracteres especiais.

### 2.2 Remoção de tweets duplicados

### 2.4 Tokenization

Para prosseguir com as análises de interesse, foi realizado o processo chamado de "tokenization", que no português pode ser traduzido como "tokenização" ou "segmentação". Esse processo consiste no particionamento dos textos das publicações em vários outros registros contendo cada uma das suas palavras componentes. Ou seja, se uma publicação tem 5 palavras, cada palavra passará a contar como uma nova linha na base de dados, porém contendo o registro do seu tweet de origem. Esta é uma etapa de muita importância para avaliar as informações das palavras componentes nas etapas seguintes.

A figura abaixo traz um exemplo de tokenization aplicado a uma frase.




```r
data_token<-data %>%  unnest_tokens(word, text)
dim(data_token)
```

```r
head(data_token)  ## ver a estrutura geral dos novos dados
```

### 2.5 Remoção de stop words

Uma etapa importante no processamento de dados textuais consiste na remoção de palavras que não contribuem com análise de conteúdo, tais como preposições, conjunções e artigos. Estas palavras são chamadas de "stop-words" e são registradas em dicionários, que são tomados como base para a remoção das palavras nos conteúdos textuais. Assim, os dados de interesse são cruzados com estes dicionários, de forma a remover aquelas palavras identificadas como "stop-words".

##### 2.5.1 Remoção de RT, Amazon e Rainforest

Antes da remoção das Stop-words, foi realizada a remoção das palavras "RT", que faz referência a uma republicação e das palavras que são fruto da busca ("Amazon" e "rainforest"). Além disso, foi feita a padronização do termo "fires" e "fire". 

```r
data_token<-data_token %>%
filter(., word!="rt") %>%
filter(., word!="amazon") %>%
filter(.,word!="rainforest")

data_token$word[data_token$word=="fires"]<-"fire"
dim(data_token)
```

##### 2.5.2 Stop words

Agora sim a remoção das Stop-words.

```r
data(stop_words)  ### carregamento de uma base de dados com as stop-words

#### realização do antijoin pra remover estas palavras
tidy_text <- data_token %>%
  anti_join(stop_words)
head(tidy_text)
dim(tidy_text)
```

### 3.5 Análise de regras de associação

Com base nos dados de ocorrência das palavras nas publicações, também foi realizada a análise de regras de associação (ARA), com o objetivo de avaliar se existem relações consistentes entre palavras nas publicações. A análise foi feita utilizando como critérios de seleção:

- suporte igual ou superior a 0.01875117 (1.875117 %), que corresponde à ocorrência da espécie no LER em pelo menos 1000 publicações; 

- confiança de 0,8 para a regra, em que as palavras em LDR devem estar relacionadas com as palavras do LER em pelo menos 80% de suas ocorrências;

- valor de p de Fisher significativo (<0,05). 

O fluxo de análise está apresentado abaixo.



```r
## carregamento de pacotes específicos
library(arules)  
library(arulesViz)

## exportação de um objeto no formato desejado para análise
write.csv(select(tidy_text,tweet_id,word),"word_tweet.csv")


########### leitura deste arquivo de transações
trans = read.transactions("word_tweet.csv", h=T, sep = ",",format = "single",cols=c("tweet_id","word"))

# Obtenção das regas de interesse
rules<- apriori(trans, parameter = list(supp = 0.01875117, conf = 0.8, target="rules", minlen=2,maxlen=2))

########## Obter medidas de interesse para as regras
measures<-interestMeasure(rules,
                               measure=c("coverage","fishersExactTest"),
                               transactions = trans)

############ Adicionar a medida de interesse ao objeto de regras
rules@quality$fisher_p<-measures$fishersExactTest

rules
```

```r
## Resultados gerais das regras
rules %>%
  sort(., by = "lift") %>%           
  head(., n =75 ) %>%                       
  inspect(.)
```

```r
######### fazer um plot para as regras obtidas


plot(rules, method="graph", main="Principais associções",
     alpha=1,cex=1,labelCol="#000000",precision=3, engine="igraph",arrowSize=1)
```

       Análise resultou em 75 regras entre pares de termos, compostas por 30 palavras que se relacionam entre si. Parte destas palavras estão presentes nas 20 mais importantes apresentadas anteriormente (como “carbono” e “fire”), enquanto outras não foram citadas anteriormente (como “ice” e “heatwaves”). Dentro do contexto avaliado de percepção ambiental, chama a atenção dentre as regras obtidas as associações consistentes entre “unnatural” e “heatwaves”, “dying” e “fire”; “emitts” e “absorbs”; “peoples” e “indigenous”; “carbon” e “dioxide” e “absorb”. Todas estas associações indicam que a ocorrência de uma palavra na publicação está fortemente relacionada a ocorrência da outra, em uma relação forte de coexistência. Também podem ser observadas outras associações consistentes de acordo com os critérios considerados, mas que não são relevantes dentro do contexto avaliado de percepção ambiental sobre a região amazônica.
